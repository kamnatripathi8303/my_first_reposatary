```python
# ðŸ“Š Sales Prediction Using Python

# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


# Step 2: Create a demo dataset
data = {
    'TV': [230.1, 44.5, 17.2, 151.5, 180.8, 8.7, 57.5, 120.2, 8.6, 199.8],
    'Radio': [37.8, 39.3, 45.9, 41.3, 10.8, 48.9, 32.8, 19.6, 2.1, 2.6],
    'Newspaper': [69.2, 45.1, 69.3, 58.5, 58.4, 75.0, 23.5, 11.6, 1.0, 21.2],
    'Sales': [22.1, 10.4, 9.3, 18.5, 12.9, 7.2, 11.8, 13.2, 4.8, 10.6]
}



df = pd.DataFrame(data)


# Step 3: Display dataset
print("ðŸ“‹ Dataset Preview:")
print(df.head())

# Step 4: Define features (X) and target (y)
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']


# Step 5: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 6: Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)


# Step 7: Make predictions
y_pred = model.predict(X_test)


# Step 8: Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nðŸ“ˆ Model Evaluation:")
print(f"Mean Squared Error: {mse:.2f}")
print(f"RÂ² Score: {r2:.2f}")


# Step 9: Visualize actual vs predicted sales
plt.scatter(y_test, y_pred, color='blue')
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted Sales")
plt.show()


# Step 10: Predict new values (example)
new_data = pd.DataFrame({'TV': [150], 'Radio': [35], 'Newspaper': [50]})
predicted_sales = model.predict(new_data)
print(f"\nðŸ§® Predicted Sales for new data: {predicted_sales[0]:.2f}")
```
